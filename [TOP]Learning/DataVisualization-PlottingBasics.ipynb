```python
import pandas as pd

fires=pd.read_csv("data/forestfires.csv")
fires.head(5)
```

```python
plt.scatter(fires["wind"], fires["area"])
plt.show()

plt.scatter(fires["temp"], fires["area"])
plt.show()
```

```python
fires=fires.sort_values(["temp"])
plt.plot(fires["temp"], fires["area"])
plt.title("How temp affects area")
plt.xlabel("Temp")
plt.ylabel("Area")
plt.show()
```
