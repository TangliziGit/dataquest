In this course, we used .hist(), .boxplot(), .scatter() to visualize the dataset.

```python
import pandas as pd

grad=pd.read_csv('data/recent-grads.csv')
grad.head(5)
```

```python
import matplotlib.pyplot as plt

grad.hist(column=['Median', 'Sample_size'], layout=(2, 1), bins=50)
```

 reference:<http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.hist.html#pandas.DataFrame.hist>

```python
# How to use boxplot
# use 'by' as domain
grad[['Sample_size', 'Major_category']].boxplot(by='Major_category')
plt.xticks(rotation=90)
```

箱型图包含六个数据节点，分别计算出他的最大值，上四分位数，中位数，下四分位数，最小值，还有一个异常值。

```python
grad[['Total', 'Major_category']].boxplot(by='Major_category')
plt.xticks(rotation=90)
```

```python
plt.scatter(grad['Unemployment_rate'], grad['Median'], color='red')
plt.scatter(grad['ShareWomen'], grad['Median'], color='blue')
plt.title('Unemployment_rate v.s. ShareWomen')
plt.show()
```
